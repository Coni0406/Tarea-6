```python
# Literal 1
import numpy as np
import matplotlib.pyplot as plt

# Función original f(x) = 1/25 * x^2 + 1
def f1(x):
    return 1/25 * x**2 + 1

# Aproximación de Taylor de f(x) alrededor de x0 = 0 (hasta el segundo orden)
def taylor_f1(x):
    return 1 + (1/25) * x**2

# Generar puntos para graficar
x_vals = np.linspace(-2, 2, 400)

# Funciones aproximadas usando la serie de Taylor
taylor_approx_f1 = taylor_f1(x_vals)

# Graficar
plt.figure(figsize=(8, 6))
plt.plot(x_vals, f1(x_vals), label="f(x) = 1/25 * x^2 + 1", color="blue")
plt.plot(x_vals, taylor_approx_f1, label="Aproximación de Taylor", linestyle='--', color="red")
plt.title("Aproximación de Taylor para f(x) = 1/25 * x^2 + 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()

# Imprimir las fórmulas utilizadas
print("Serie de Taylor para f(x) = 1/25 * x^2 + 1 alrededor de x0 = 0:")
print("f(x) ≈ 1 + (1/25) * x^2")

# Polinomio de Lagrange con dos puntos: x0 = 0 y x1 = 1
def lagrange_polynomial(x):
    x0, x1 = 0, 1
    f_at_x0, f_at_x1 = f1(x0), f1(x1)  # Cambié el nombre de la variable f1 a f_at_x0, f_at_x1
    L0 = (x - x1) / (x0 - x1)
    L1 = (x - x0) / (x1 - x0)
    return f_at_x0 * L0 + f_at_x1 * L1

# Graficar el polinomio de Lagrange
lagrange_approx_f1 = lagrange_polynomial(x_vals)
plt.figure(figsize=(8, 6))
plt.plot(x_vals, f1(x_vals), label="f(x) = 1/25 * x^2 + 1", color="blue")
plt.plot(x_vals, lagrange_approx_f1, label="Polinomio de Lagrange", linestyle='--', color="green")
plt.title("Polinomio de Lagrange para f(x) = 1/25 * x^2 + 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Literal 2
# Función original f(x) = arctan(x)
def f2(x):
    return np.arctan(x)

# Aproximación de Taylor de f(x) alrededor de x0 = 1 (hasta el tercer orden)
def taylor_f2(x):
    return np.pi/4 + (1/2)*(x - 1) - (1/8)*(x - 1)**2 - (1/48)*(x - 1)**3

# Generar puntos para graficar
x_vals_f2 = np.linspace(0, 2, 400)

# Funciones aproximadas usando la serie de Taylor
taylor_approx_f2 = taylor_f2(x_vals_f2)

# Graficar
plt.figure(figsize=(8, 6))
plt.plot(x_vals_f2, f2(x_vals_f2), label="f(x) = arctan(x)", color="blue")
plt.plot(x_vals_f2, taylor_approx_f2, label="Aproximación de Taylor", linestyle='--', color="red")
plt.title("Aproximación de Taylor para f(x) = arctan(x) alrededor de x0 = 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()

# Imprimir las fórmulas utilizadas
print("\nSerie de Taylor para f(x) = arctan(x) alrededor de x0 = 1:")
print("f(x) ≈ π/4 + (1/2)(x - 1) - (1/8)(x - 1)^2 - (1/48)(x - 1)^3")

# Polinomio de Lagrange con dos puntos: x0 = 1 y x1 = 1.1
def lagrange_polynomial_f2(x):
    x0, x1 = 1, 1.1
    f0, f1 = f2(x0), f2(x1)
    L0 = (x - x1) / (x0 - x1)
    L1 = (x - x0) / (x1 - x0)
    return f0 * L0 + f1 * L1

# Graficar el polinomio de Lagrange para f(x) = arctan(x)
lagrange_approx_f2 = lagrange_polynomial_f2(x_vals_f2)
plt.figure(figsize=(8, 6))
plt.plot(x_vals_f2, f2(x_vals_f2), label="f(x) = arctan(x)", color="blue")
plt.plot(x_vals_f2, lagrange_approx_f2, label="Polinomio de Lagrange", linestyle='--', color="green")
plt.title("Polinomio de Lagrange para f(x) = arctan(x) alrededor de x0 = 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
